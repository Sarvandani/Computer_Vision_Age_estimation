```python
# Install DeepFace
!pip install deepface

# Import libraries
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import urllib.request
import numpy as np

# Image URL from PostImage
url = 'https://i.postimg.cc/j5Zm8L3R/axxxaaa.png'

# Download the image
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_COLOR)

# Convert BGR to RGB for displaying and processing
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Perform age estimation
analysis = DeepFace.analyze(image_rgb, actions=['age'])

# Print estimated age(s)
if isinstance(analysis, list):  # multiple faces
    for i, face in enumerate(analysis):
        print(f"Face {i+1} estimated age: {face['age']}")
else:  # single face
    print(f"Estimated age: {analysis['age']}")
```
